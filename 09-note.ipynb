```python
import random
while True:
 rand_num = random.randint(1,6)
 print(rand_num)
 if rand_num == 4:
    break
 print('.')
```

4がでるまで.をだす while Trueで無限ループ breakで無限ループを止める　
ここでの条件は4がでたらとめる

```python
import random
while True:
 rand_num = random.randint(1,6)
 print(rand_num)
 if rand_num == 4:
    continue
 print('.')
```

continueで無限ループさせるだが4の後ろにはピリオドが付かない　条件がでても無限ループは続く
